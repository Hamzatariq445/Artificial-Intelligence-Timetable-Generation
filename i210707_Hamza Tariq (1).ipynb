# Genetic Algo

```python
# Function for calculate_fitness
def calculate_fitness(chromosome, rooms, courses, sections):
    fitness = 0
    # checks if the there exist list with this key then ok else makes the new empty list despit of giving error
    used_slots = defaultdict(list)
    used_professors = defaultdict(list)
    used_rooms = defaultdict(list)
    used_course_sections = defaultdict(list)

    for gene in chromosome:
        course_id, course_type, section, section_strength, professor_name, day1, timeslot1, room1, room_size1, day2, timeslot2, room2, room_size2 = gene

        # Hard constraints
        # 1. Classes can only be scheduled in free classrooms
        if (day1, timeslot1, room1) in used_rooms or (day2, timeslot2, room2) in used_rooms:
            fitness += 1

        # 2. A classroom should be big enough to accommodate the section
        if courses[course_id][2] > room_size1 or courses[course_id][2] > room_size2:
            fitness += 1

        # 3. A professor should not be assigned two different lectures at the same time
        if (day1, timeslot1) in used_professors[professor_name] or (day2, timeslot2) in used_professors[professor_name]:
            fitness += 1

        # 4. The same section cannot be assigned to two different rooms at the same time
        if (day1, timeslot1) in used_slots[section] or (day2, timeslot2) in used_slots[section]:
            fitness += 1

        # 5. A room cannot be assigned for two different sections at the same time
        if (day1, timeslot1, room1) in used_rooms or (day2, timeslot2, room2) in used_rooms:
            fitness += 1

        # 6. No professor can teach more than 3 courses
        if len(used_professors[professor_name]) >= 3:
            fitness += 1

        # 7. No section can have more than 5 courses in a semester
        if len(used_slots[section]) >= 5:
            fitness += 1

        # 8. Each course would have two lectures per week not on the same or adjacent days
        if course_type == "Theory" and (abs(day1 - day2) < 2 or (day1 == day2 and abs(timeslot1 - timeslot2) < 2)):
            fitness += 1

        # 9. Lab lectures should be conducted in two consecutive slots
        if course_type == "Lab" and timeslot2 != timeslot1 + 1:
            fitness += 1

        # 10. 15 mins breaks allowed between consecutive classes to ensure sufficient time for transitions
        for day, slot, room in used_rooms:
            if (day == day1 and abs(slot - timeslot1) == 1 and room == room1) or (day == day2 and abs(slot - timeslot2) == 1 and room == room2):
                fitness += 1

        # 11. Each section has only 1 lab of the specified course in a week
        if course_type == "Lab":
            for day, slot in used_slots[section]:
                if day == day1 or (day == day2 and slot == timeslot2 - 1):
                    fitness += 1
                    break

        # 12. Ensure day1 and day2 for lab courses are the same
        if course_type == "Lab" and day1 != day2:
            fitness += 1

        # 13. The same course of the same section cannot be assigned to the same time slot
        course_section = (course_id, section)
        if (day1, timeslot1) in used_course_sections[course_section] or (day2, timeslot2) in used_course_sections[course_section]:
            fitness += 1

        # Update used resources
        used_slots[section].append((day1, timeslot1))
        used_slots[section].append((day2, timeslot2))
        used_professors[professor_name].append((day1, timeslot1))
        used_professors[professor_name].append((day2, timeslot2))
        used_rooms[(day1, timeslot1, room1)].append(section)
        used_rooms[(day2, timeslot2, room2)].append(section)
        used_course_sections[course_section].append((day1, timeslot1))
        used_course_sections[course_section].append((day2, timeslot2))

    # Soft constraints
    # 1. All the theory classes should be taught in the morning session and all the lab sessions should be done in the afternoon session
    for gene in chromosome:
        course_id, course_type, _, _, _, day1, timeslot1, _, _, day2, timeslot2, _, _ = gene
        if course_type == "Theory" and (timeslot1 >= AFTERNOON_SESSION_START or timeslot2 >= AFTERNOON_SESSION_START):
            fitness += 1
        elif course_type == "Lab" and (timeslot1 < AFTERNOON_SESSION_START or timeslot2 < AFTERNOON_SESSION_START):
            fitness += 1

    # 2. Teachers/students may be facilitated by minimizing the number of floors they have to traverse
    for gene_index in range(len(chromosome) - 1):
        if day1 == day2:
            room1name = room_names[room1]
            room2name= room_names[room2]
            room1_floor = int(room1name.split('-')[1])
            room2_floor = int(room2name.split('-')[1])
            if abs(room1_floor - room2_floor) > 1:
                fitness += 1

    # 3. A class should be held in the same classroom across the whole week
    for gene in chromosome:
        _, _, _, _, _, _, _, room1, _, _, _, room2, _ = gene
        if room1 != room2:
            fitness += 1

    # 4. Teachers may prefer longer blocks of continuous teaching time to minimize interruptions and maximize productivity except when the courses are different
    # Implement this constraint here

    return fitness


# Genetic algorithm
def genetic_algorithm(rooms, courses, sections):
  # population generation
    population = generate_initial_population(rooms, courses, sections)
  # selection bassed on fitness calculation
    best_chromosome = min(population, key=lambda c: calculate_fitness(c, rooms, courses, sections))
  # fitness calculation of best chorormosome
    best_fitness = calculate_fitness(best_chromosome, rooms, courses, sections)

    for generation in range(MAX_GENERATIONS):
  # initializing the choromosome and performing crossover and mutation
        population = generate_new_population(population, rooms, courses, sections)
  # calculating fitness of each choromosome generated and selects minimum
        new_best_chromosome = min(population, key=lambda c: calculate_fitness(c, rooms, courses, sections))
  # calculating the fitness of best choromosome
        new_best_fitness = calculate_fitness(new_best_chromosome, rooms, courses, sections)

        if new_best_fitness < best_fitness:
            best_chromosome = new_best_chromosome
            best_fitness = new_best_fitness
            print(f"Generation {generation}: New best fitness = {best_fitness}")

    return best_chromosome



def generate_new_population(population, rooms, courses, sections):
    new_population = []
    # calculating fitness of both parents
    fitness_scores = [(calculate_fitness(chromosome, rooms, courses, sections), chromosome) for chromosome in population]
    # sorting them according to the fitness score
    sorted_population = [chromosome for _, chromosome in sorted(fitness_scores)]

    # elitism strategy
    elitism_count = int(POPULATION_SIZE * 0.1)
    # add more than 1 in list
    new_population.extend(sorted_population[:elitism_count])

    # Crossover with one parent as the best chromosome and the other as a random chromosome
    while len(new_population) < POPULATION_SIZE:
        parent1 = sorted_population[0]
        parent2 = random.choice(sorted_population)
        child1, child2 = crossover(parent1, parent2)
        mutate(child1, MUTATION_RATE)
        mutate(child2, MUTATION_RATE)
        new_population.append(child1)
        new_population.append(child2)

    return new_population[:POPULATION_SIZE]


def mutate(chromosome, mutation_rate):
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:
            gene = list(chromosome[i])
            # Exclude course ID, type, section, section strength
            mutation_point = random.randint(7, len(gene) - 1)
            # random rooms mutation
            if mutation_point == 7 or mutation_point == 11:
                room_index = random.randint(0, len(rooms) - 1)
            # update room Capacity
                room_capacity = rooms[room_index][1]
            # checks for the capacity according to random value
                while room_capacity != 60 and room_capacity != 120:
            # again perform mutation untill fix the capacity
                    room_index = random.randint(0, len(rooms) - 1)
                    room_capacity = rooms[room_index][1]
                gene[mutation_point] = room_index
            chromosome[i] = tuple(gene)
    return chromosome


# 1 point crossover
def crossover(parent1, parent2):
    child1 = []
    child2 = []

    # Select crossover point
    crossover_point = random.randint(1, len(parent1) - 1)

    for i in range(len(parent1)):
        if i < crossover_point:
            child1.append(parent1[i])
            child2.append(parent2[i])
        else:
          # Check if i is within the bounds of parents
            if i < len(parent2):
                child1.append(parent2[i])
            if i < len(parent1):
                child2.append(parent1[i])

    # checks for valid room and capcity if crossover point is in between them,
    for child in [child1, child2]:
        for i in range(len(child)):
            gene = list(child[i])
            # Check and fix room indices
            if gene[7] >= len(rooms):  # room1
                gene[7] = random.randint(0, len(rooms) - 1)
            if gene[11] >= len(rooms):  # room2
                gene[11] = random.randint(0, len(rooms) - 1)
            # Check and fix room capacities
            room_capacity1 = rooms[gene[7]][1]
            room_capacity2 = rooms[gene[11]][1]
            if room_capacity1 != 60 and room_capacity1 != 120:
                gene[7] = random.randint(0, len(rooms) - 1)
            if room_capacity2 != 60 and room_capacity2 != 120:
                gene[11] = random.randint(0, len(rooms) - 1)
            child[i] = tuple(gene)

    return child1, child2

```

```python

# Calculate the number of sections each professor is teaching
professor_section_counts = defaultdict(int)
for gene in best_chromosome:
    _, _, section, _, professor_name, _, _, _, _, _, _, _, _ = gene
    professor_section_counts[professor_name] += 1

# Create a new sheet to display the professor names and their section counts
sheet = workbook.create_sheet(title="Professor Section Counts")
sheet.append(["Professor Name", "Number of Sections"])
for professor_name, section_count in sorted(professor_section_counts.items(), key=lambda x: x[1], reverse=True):
    sheet.append([professor_name, section_count])

def add_timetable_to_sheet(sheet, day_name, timetable, current_day_index):
    sheet.title = day_name
    sheet.append(["Course", "Type", "Section", "Section Strength", "Professor", "Day", "Timeslot", "Room", "Room Size"])
    for gene in timetable:
        course_id, course_type, section, section_strength, professor_name, day1, timeslot1, room1, room_size1, day2, timeslot2, room2, room_size2 = gene
        if (day1 == current_day_index or day2 == current_day_index) and (0 <= current_day_index < DAYS_PER_WEEK):
            if day1 == current_day_index:
                time1 = timeslot_to_time(timeslot1)
                if course_type == "Lab":
                    room_name1 = lab_names.get(room1, "Unknown")
                    course_name = lab_Courses_names.get(course_id, "Unknown")
                else:
                    room_name1 = room_names.get(room1, "Unknown")
                    course_name = courses_names.get(course_id, "Unknown")
                section_label = section_names.get(section, "Unknown")
                sheet.append([course_name, course_type, section_label, section_strength, professor_name, day_name, time1, room_name1, room_size1])
            if day2 == current_day_index:
                time2 = timeslot_to_time(timeslot2)
                if course_type == "Lab":
                    room_name2 = lab_names.get(room2, "Unknown")
                    course_name = lab_Courses_names.get(course_id, "Unknown")
                else:
                    room_name2 = room_names.get(room2, "Unknown")
                    course_name = courses_names.get(course_id, "Unknown")
                section_label = section_names.get(section, "Unknown")
                sheet.append([course_name, course_type, section_label, section_strength, professor_name, day_name, time2, room_name2, room_size2])

# Write each day's timetable to a separate sheet
for day_index in range(DAYS_PER_WEEK):
    day_name = day_names[day_index]
    sheet = workbook.create_sheet()
    add_timetable_to_sheet(sheet, day_name, best_chromosome, day_index)


# Save the workbook to a file
workbook.save("timetable.xlsx")

def create_timetable_grid(sheet, day_name, timetable):
    # Create headers for the grid
    sheet.append([""] + [timeslot_to_time(timeslot) for timeslot in range(TIMESLOTS_PER_DAY)])

    # Add data to the grid
    for room in range(NUM_ROOMS):
        row = [room_names.get(room, "Unknown")]
        for timeslot in range(TIMESLOTS_PER_DAY):
            room_data = []
            for gene in timetable:
                course_id, course_type, section, _, professor_name, day1, timeslot1, room1, _, day2, timeslot2, room2, _ = gene
                if (day_name == day_names[day1] and timeslot1 == timeslot and room1 == room) or (day_name == day_names[day2] and timeslot2 == timeslot and room2 == room):
                    section_label = section_names.get(section, "Unknown")
                    room_data.append(f"{course_id}/{section_label}/{professor_name}")
            row.append("\n".join(room_data))
        sheet.append(row)

# Write the timetable grid for each day in separate sheets
for day_index in range(DAYS_PER_WEEK):
    day_name = day_names[day_index]
    sheet = workbook.create_sheet(title=f"{day_name} Grid")
    create_timetable_grid(sheet, day_name, best_chromosome)

# Save the workbook to a file
workbook.save("timetable.xlsx")

print("SuccessFully Written in Excel.")
```
